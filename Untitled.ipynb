```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import xgboost as xgb
import catboost as cb

# 1. 数据准备与特征工程
df = pd.read_csv('Super_Cleaned_Concrete_Data.csv')
df = df.drop(columns=['SF'], errors='ignore')
df['TOTAL_BINDER'] = df['PC'] + df['FA'] + df['SC']
df['w/b'] = df['WATER'] / df['TOTAL_BINDER']
agg_sum = df['FAGG'] + df['CAGG']
df['b/a'] = df['TOTAL_BINDER'] / agg_sum
df['SCM%'] = (df['FA'] + df['SC']) / df['TOTAL_BINDER']
df['CAGG%'] = df['CAGG'] / agg_sum
df['FAGG%'] = df['FAGG'] / agg_sum
df['PC%'] = df['PC'] / df['TOTAL_BINDER']
df['FA%'] = df['FA'] / df['TOTAL_BINDER']
df['SC%'] = df['SC'] / df['TOTAL_BINDER']

base_features = [col for col in df.columns if col not in ['7day', '28day', '56day']]
targets = ['7day', '28day', '56day']

# 2. 定义寻优函数
def get_best_model(model_name, X, y):
    if model_name == 'RF':
        params = {'n_estimators': [100, 200], 'max_depth': [None, 10, 20]}
        search = RandomizedSearchCV(RandomForestRegressor(), params, n_iter=3, cv=3, n_jobs=-1)
    elif model_name == 'XGBoost':
        params = {'learning_rate': [0.01, 0.1], 'max_depth': [3, 6]}
        search = RandomizedSearchCV(xgb.XGBRegressor(), params, n_iter=3, cv=3, n_jobs=-1)
    elif model_name == 'CatBoost':
        params = {'depth': [4, 6], 'learning_rate': [0.01, 0.1], 'iterations': [500]}
        search = RandomizedSearchCV(cb.CatBoostRegressor(logging_level='Silent'), params, n_iter=3, cv=3, n_jobs=-1)
    elif model_name == 'MLP':
        pipe = Pipeline([('scaler', StandardScaler()), ('mlp', MLPRegressor(max_iter=1000))])
        params = {'mlp__hidden_layer_sizes': [(50,), (100, 50)]}
        search = RandomizedSearchCV(pipe, params, n_iter=2, cv=3, n_jobs=-1)
    
    search.fit(X, y)
    return search.best_estimator_

# 3. 循环 3 次实验
results = []
for i in range(3):
    print(f"执行第 {i+1}/3 次随机实验 (带参数寻优)...")
    seed = 42 + i
    
    # A. 基准模型 (独立预测)
    for name in ['RF', 'XGBoost', 'CatBoost', 'MLP']:
        for target in targets:
            t_df = df.dropna(subset=[target])
            X_tr, X_te, y_tr, y_te = train_test_split(t_df[base_features], t_df[target], test_size=0.2, random_state=seed)
            model = get_best_model(name, X_tr, y_tr)
            results.append({'Model': name, 'Age': target, 'R2': r2_score(y_te, model.predict(X_te))})
            
    # B. 链式 CatBoost (寻优版)
    chained_models = {}
    _, test_df = train_test_split(df.dropna(subset=targets), test_size=0.2, random_state=seed)
    
    # 每一轮针对当前的 target 进行独立的参数寻优
    for target in targets:
        # 1. 准备数据：如果有前置龄期，将其作为输入特征
        t_df = df.dropna(subset=[target]).copy()
        prev_target = {'28day':'7day', '56day':'28day'}
        if target in prev_target:
            t_df = t_df.dropna(subset=[prev_target[target]])
        
        # 确定当前的输入特征
        curr_feat = base_features.copy()
        if target in prev_target:
            curr_feat.append(prev_target[target])
        
        X_tr, X_te, y_tr, y_te = train_test_split(t_df[curr_feat], t_df[target], test_size=0.2, random_state=seed)
        
        # 2. 关键：对每个阶段的模型进行参数寻优
        print(f"  正在为 {target} 阶段模型进行参数搜索...")
        best_cat = get_best_model('CatBoost', X_tr, y_tr)
        chained_models[target] = best_cat
        
    # 3. 链式推理评估
    # 确保使用测试集上的特征和前置预测结果
    pred_7 = chained_models['7day'].predict(test_df[base_features])
    
    # 构造28d输入
    X_28 = test_df[base_features].copy()
    X_28['7day'] = pred_7
    pred_28 = chained_models['28day'].predict(X_28)
    
    # 构造56d输入
    X_56 = test_df[base_features].copy()
    X_56['28day'] = pred_28
    pred_56 = chained_models['56day'].predict(X_56)
    
    # 4. 存入结果
    for t, p in zip(targets, [pred_7, pred_28, pred_56]):
        results.append({'Model': 'CatBoost-Chain', 'Age': t, 'R2': r2_score(test_df[t], p)})


# 4. 导出 CSV
results_df = pd.DataFrame(results)
results_df.to_csv('model_comparison_results.csv', index=False)
print("\n✅ 所有结果已成功导出至 model_comparison_results.csv")
```
